```python
t_statistic, p_value  =  stats.ttest_ind(Mrs_Smith,Mrs_Jones)
print('P Value %1.7f' % p_value)    
print('t stats Value %1.7f' % t_statistic)  
```

Q.2B). State the null and alternative hypothesis for any one of the test mention in the question number one 

```python
## We are performing two sample ttest for gender vs Monthly charge
dfmale = df[(df['gender'] == 'Male')]
male_monthly = np.array(dfmale.MonthlyCharges)
male_monthly
print(len(male_monthly))
```

```python
dfemale = df[(df['gender'] == 'Female')]
female_monthly = np.array(dfemale.MonthlyCharges)
female_monthly
print(len(female_monthly))
```

As we can see the both the sizes of male_monthly and female_monthly are not same we can go with two sample t-test for unpaired data.

```python
## H0 = mean_male_monthly = mean_female_monthly
## Ha = mean_male_monthly != mean_female_monthly
```

```python
t_statistics, p_value = stats.ttest_ind(male_monthly, female_monthly)
t_statistics, p_value
```

From the above results of the test we have found that P-value of test is 0.22 which is greater than Alpha value 0.05.

So the statistical decision is to accept the null hypothesis at 5% level of significance.

Which also mean that there is a no difference between means of male_monthly and female_monthly. 

Q.2C). Now perform any two required statistical test on the data to determine what are the significant features in the data with respect to target variable (95% C.I)? (4 Marks)

First we perform two sample two way chisquare test with respect to target variable

1st test we prefer to do a chi square test between churn and gender

```python
data=df.groupby(['gender','Churn']).agg({'Churn':'count'})
data
```

Null hypothesis:  𝐻0 : There is no difference in churn of male and female

Alternative hypothesis:  𝐻𝐴 : There is significant difference in churn of male and female

```python
a = pd.array(data.Churn[0:2])
b = pd.array(data.Churn[2:4])
```

```python
quality_array = np.array([a,b])
chi_sq_Stat, p_value, deg_freedom, exp_freq = stats.chi2_contingency(quality_array)

print('Chi-square statistic %3.5f P value %1.6f Degrees of freedom %d' %(chi_sq_Stat, p_value,deg_freedom))
```

As from the above test result we can say that  p value is 0.48 which is > 0.05 so we accept the null hypothesis.

So, we conclude that there is a no difference in the churn of male and female..


2nd test we prefer to chi square test between churn and contract

```python
data1 =df.groupby(['Contract','Churn']).agg({'Churn':'count'})
data1
```

Null hypothesis:  𝐻0 : There is no difference in churn and contract

Alternative hypothesis:  𝐻𝐴 : There is a significant difference in churn and contract

```python
c = pd.array(data1.Churn[0:2])
d = pd.array(data1.Churn[2:4])
e = pd.array(data1.Churn[4:6])
```

```python
churn_array = np.array([c,d,e])
chi_sq_Stat, p_value, deg_freedom, exp_freq = stats.chi2_contingency(churn_array)

print('Chi-square statistic %3.5f P value %1.10f Degrees of freedom %d' %(chi_sq_Stat, p_value,deg_freedom))
```

As from the above test result we can say that  p value is 0.00 which is < 0.05 so we fail to accept the null hypothesis.

So, we conclude that there is a significant difference in the churn of contract.
